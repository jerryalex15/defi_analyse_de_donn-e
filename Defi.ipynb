# Défi Analyse de Données

```python
import pandas as pd
from sklearn.neural_network import MLPClassifier
```

```python
training_class = pd.read_csv("idc-challenge-2024/training_classes.csv")
training_data = pd.read_csv("idc-challenge-2024/training_data.csv")
test_data = pd.read_csv("idc-challenge-2024/test_data.csv")
test_data
```

```python
#Enlève les données  catégorielles (non numérique)
t_data = training_data.select_dtypes(include=['number'])
t_class = training_class.select_dtypes(include=['number'])
testData = test_data.select_dtypes(include=['number'])
testData
```

```python
clf_neural = MLPClassifier()
clf_neural.fit(t_data, training_class)
```

```python
yproba_neural = clf_neural.predict_proba(testData)
print(clf_neural.classes_)
#yproba_neural
```

```python
df = pd.DataFrame({'Id': range(1, len(yproba_neural)+1), 'Predicted': yproba_neural[:,1]})
df
```

```python
df.to_csv('data.csv', sep=',', index=False)
```
