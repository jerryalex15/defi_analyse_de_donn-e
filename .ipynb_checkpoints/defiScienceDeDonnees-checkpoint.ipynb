```python
# Récupère les colonnes qui contiennent des données catégoriques.
categoricals = Xtrain.select_dtypes(include=['object']).columns.tolist()
# Récupère les colonnes qui contiennent des données numériques.
numericals = Xtrain.select_dtypes(include=["number"]).columns.tolist() 

preprocessor = ColumnTransformer([
    ('encoder', OneHotEncoder(), categoricals),
    ('scaler', StandardScaler(), numericals)
])
```

```python
# Define the pipeline
def find_k_best_feature(K, classifier):
    pipeline = Pipeline([
        ('process', preprocessor),
        ('feature_selection', SelectKBest(score_func=f_classif, k=K)),  # Select top 2 features  , essayer sur 3, 4 , 5 ; ... et de tracer
        ('classification', classifier)                                  # On prend ce qui donne la meilleur accuracy
    ])

    # Définir la stratégie de validation croisée
    kfold = KFold(n_splits=5, shuffle=True, random_state=42)
    
    # Utiliser cross_val_score pour évaluer le pipeline avec la validation croisée
    scores = cross_val_score(pipeline, Xtrain, Ytrain.values.ravel(), cv=kfold)
    
    # Evaluate the pipeline
    # evaluation = pipeline.score(X_test, y_test).

    return scores.mean()
```

Je vais considérer les meilleurs attributs propre à chacune de ces 3 classifieurs et de l'éxperimenter. je m'interesse maintenant aux hyperparamètres de chaque classifiers et de rétenir ceux qui performent le mieux.

Commençons par le classifieur SVM:

```python
# Définir les hyperparamètres à rechercher pour SVC
parameters = {
    'classifier__kernel': ['linear', 'poly'],
    'classifier__gamma': [0.1, 0.01],
}
print ('Création du pipeline \n')
pipeline = Pipeline([
    ('process', preprocessor),
    ('feature_selection', SelectKBest(score_func=f_classif, k=17)), # k = 17 , c'est le nombre des meilleurs attibuts pour le SVM
    ('classifier', SVC(C=0.1, probability=True))                    
])

# Recherche par grille pour trouver les meilleurs hyperparamètres
grid_search = GridSearchCV(estimator=pipeline, 
                           param_grid=parameters,
                           scoring='accuracy',
                           cv=5,
                           n_jobs=1,
                           #iid=True,
                           return_train_score=True)
grid_search.fit(X_train, y_train.values.ravel())

# Afficher les meilleurs hyperparamètres
print("Meilleurs hyperparamètres :", grid_search.best_params_)

# Utiliser le meilleur modèle pour prédire
y_pred = grid_search.predict_proba(X_test)

# Évaluer les performances du modèle
accuracy = grid_search.score(X_test, y_test.values.ravel())
print("Exactitude du modèle avec les meilleurs hyperparamètres :", accuracy)
```
